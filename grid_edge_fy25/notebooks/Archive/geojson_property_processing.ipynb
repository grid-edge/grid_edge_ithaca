```python
# install libraries
import os
import json
import numpy as np
import pandas as pd

from datetime import datetime
from pathlib import Path
from collections import defaultdict
```

```python
## Load NY-Dryden GeoJSON
current_path = Path(os.getcwd())
data_path = current_path / '..'/ 'data' 
geojson_path = data_path / 'Main Street Dryden.geojson'

## Load NY-Dryden Property Data
property_path = data_path / 'Dryden Parcels.json'

print('done')
```

```python
# read geojson

# Load GeoJSON file
with open(geojson_path, "r") as f:
    geojson = json.load(f)

# Loop through features
for feature in geojson.get("features", []):
    geometry = feature.get("geometry", {})
    properties = feature.get("properties", {})
    addr_housenumber = properties.get("addr:housenumber")
    addr_street = properties.get("addr:street", "").lower()
    building = properties.get("building", "").lower()

    print("House Number:", addr_housenumber)
    print("Street:", addr_street)
    print("Building:", building)
    print("Geometry:", geometry.get("type"), geometry.get("coordinates"))
    print("---")
```

```python
# read property data

# Load property file
with open(property_path, "r") as f:
    property_json = json.load(f)

# loop through IDs

for parcel_id, properties in property_json.items():

    address = properties.get("address", "").lower()
    property_type = properties.get("propertyType", "").lower()
    street_name = properties.get("streetName", "").lower()

    print(f"Parcel ID: {parcel_id}")
    print(f"  Address: {address}")
    print(f"  Property Type: {property_type}")
    print(f"  Street Name: {street_name}")
    print()
```
